# 绘制简单的折线图 #

```python
import matplotlib.pyplot as plt
%matplotlib inline 
squares=[1,4,9,16,25]
print(squares)
plt.plot(squares)
plt.show()
```

# 修改标签文字和线条粗细 #

```python
squares=[1,4,9,16,25]
plt.plot(squares,linewidth=5)
#设置图表标题 并给坐标轴加上标签
plt.title('square numbers',fontsize=24)
plt.xlabel('values',fontsize=14)
plt.ylabel('square of value',fontsize=14)
#设置刻度标记的大小
plt.tick_params(axis='both',labelsize=14)
plt.show()
```

# 校正图形 #

```python
input_values=[1,2,3,4,5]
squares=[1,4,9,16,25]
plt.plot(input_values,squares,linewidth=5)
#设置图表标题 并给坐标轴加上标签
plt.title('square numbers',fontsize=24)
plt.xlabel('values',fontsize=14)
plt.ylabel('square of value',fontsize=14)
#设置刻度标记的大小
plt.tick_params(axis='both',labelsize=14)
plt.show()
```


# 使用scatter（）绘制散点图并设置其样式 #

```python
plt.scatter(2,4)
plt.show()
```

```python
plt.scatter(2,4,s=200)
#设置图表标题 并给坐标轴加上标签
plt.title('square numbers',fontsize=24)
plt.xlabel('values',fontsize=14)
plt.ylabel('square of value',fontsize=14)
#设置刻度标记的大小
plt.tick_params(axis='both',which='major',labelsize=14)
plt.show()
```

# 使用scatter绘制一系列点 #

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.scatter(x,y,s=100)
#设置图表标题 并给坐标轴加上标签
plt.title('square numbers',fontsize=24)
plt.xlabel('values',fontsize=14)
plt.ylabel('square of value',fontsize=14)
#设置刻度标记的大小
plt.tick_params(axis='both',which='major',labelsize=14)
plt.show()
```

# 自动计算数据 #

```python
x=list(range(1,1001))
y=[x**2 for x in x]
plt.scatter(x,y,s=40)
#设置图表标题 并给坐标轴加上标签
plt.title('square numbers',fontsize=24)
plt.xlabel('values',fontsize=14)
plt.ylabel('square of value',fontsize=14)
#设置刻度标记的大小
plt.tick_params(axis='both',which='major',labelsize=14)
plt.axis([0,1100,0,1100000])
plt.show()
```

# 自定义颜色 #

```python
x=list(range(1,1001))
y=[x**2 for x in x]
plt.scatter(x,y,c='red',edgecolor='none',s=40)
#设置图表标题 并给坐标轴加上标签
plt.title('square numbers',fontsize=24)
plt.xlabel('values',fontsize=14)
plt.ylabel('square of value',fontsize=14)
#设置刻度标记的大小
plt.tick_params(axis='both',which='major',labelsize=14)
plt.axis([0,1100,0,1100000])
plt.show()
```

# 使用颜色映射 #

```python
x=list(range(1,1001))
y=[x**2 for x in x]
plt.scatter(x,y,c=y,cmap=plt.cm.Blues,edgecolor='none',s=40)
#设置图表标题 并给坐标轴加上标签
plt.title('square numbers',fontsize=24)
plt.xlabel('values',fontsize=14)
plt.ylabel('square of value',fontsize=14)
#设置刻度标记的大小
plt.tick_params(axis='both',which='major',labelsize=14)
plt.axis([0,1100,0,1100000])
plt.show()
```
